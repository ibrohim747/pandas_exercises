```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
```

```python
g = np.random.default_rng(0)
month = "Sep Oct Nov Dec Jan Feb Mar Apr May Jun".split()
s = Series(g.integers(40, 60, 10), index = month)
s
```

```python
print(s.mean(), s.std())
for i in s.index:
  different = 80 - s.mean()
  s[i] = s[i] + different
```

```python
s
```

```python
s[s > s.mean() + s.std()]
```

```python
s[(s < s.mean() + s.std()) & (s > s.mean())]
```

```python
s[(s > s.mean() - s.std()) & (s < s.mean())]
```

```python
s[s < s.mean() - s.std()]
```

```python
s[(s < s.mean() - 2*s.std()) |  (s > s.mean() + 2*s.std())]
```
