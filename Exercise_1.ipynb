```python
g = np.random.default_rng(0)
s = Series(g.integers(70, 101, 10))
s
```

```python
s.index = "Sep Oct Nov Dec Jan Feb Mar Apr May Jun".split()
s
```

```python
g = np.random.default_rng(0)
month = "Sep Oct Nov Dec Jan Feb Mar Apr May Jun".split()
s = Series(g.integers(70, 101, 10), index = month)
s
```

```python
print(f"Yearly average: {s.mean()}")
```

```python
year_1 = s[:5].mean()
year_2 = s[5:].mean()
print(f"1 year average: {year_1}")
print(f"2 year average: {year_2}")
```

```python
first_half_year = s.loc["Sep": "Jan"].mean()    # loc = location   iloc = index location
second_half_year = s.loc["Feb": "Jun"].mean()
print(f"First half year average: {first_half_year}")
print(f"Second half year average: {second_half_year}")
```

```python
head_1 = s.head(5).mean()
tail_1 = s.tail(5).mean()
print(f"First 5 months average: {head_1}")
print(f"Last 5 months average: {tail_1}")
```
